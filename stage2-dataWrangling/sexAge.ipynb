```python
filepaths = [os.path.join(folder, filename) for filename in os.listdir(folder)]
filepaths
```

```python
def importVars(filepath):
    return pd.read_csv(filepath, usecols=["PcrKey", "ePatient_13", "ePatient_15"], dtype=np.int32)
```

```python
df = pd.concat(map(importVars, filepaths))
```

```python
df
```

```python
df = df.rename(columns= {'ePatient_13' : 'Gender', 'ePatient_15' : 'age'})
# female = 9906001, change to 1
df = df.replace([9906001], 1)
# male = 9906003, change to 0
df = df.replace([9906003], 0)
```

```python
df.head()
```

```python
# drop missing data rows for gender
df = df[ (df['Gender'] == 0) | (df['Gender'] == 1) ]
```

```python
# drop missing data rows for age
df = df[ (df['age'] >= 0) & (df['age'] < 150)]
```
